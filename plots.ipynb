```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.gridspec import GridSpec
```

```python
# KNN plots
# Precision, Recall, and F1-Score (KNN)
classes = {
    0: "Bladder", 
    1: "Femur-left", 
    2: "Femur-right", 
    3: "Heart", 
    4: "Kidney-left", 
    5: "Kidney-right", 
    6: "Liver", 
    7: "Lung-left", 
    8: "Lung-right", 
    9: "Pancreas", 
    10: "Spleen"
}

# base KNN data
base_knn = {
    'class': list(range(11)),
    'precision': [0.89, 0.96, 0.95, 0.81, 0.49, 0.61, 0.61, 0.98, 0.99, 0.40, 0.41],
    'recall': [0.90, 0.95, 0.95, 0.58, 0.49, 0.68, 0.73, 0.93, 0.87, 0.31, 0.51],
    'f1-score': [0.90, 0.95, 0.95, 0.68, 0.49, 0.64, 0.66, 0.95, 0.93, 0.35, 0.46]
}

# KNN with image processing and augmentation
optimized_knn = {
    'class': list(range(11)),
    'precision': [0.47, 0.65, 0.78, 0.90, 0.63, 0.76, 0.88, 0.93, 0.98, 0.56, 0.45],
    'recall': [0.67, 0.58, 0.57, 0.80, 0.61, 0.59, 0.97, 0.97, 0.87, 0.46, 0.64],
    'f1-score': [0.55, 0.61, 0.66, 0.85, 0.62, 0.66, 0.92, 0.95, 0.92, 0.50, 0.53]
}

# create figure and axis
fig, ax = plt.subplots(figsize=(15, 10))
x = np.arange(len(classes))
width = 0.1  # Width of each bar
ax.bar(x - 0.3, base_knn['precision'], width, label='Base KNN - Precision', color='#3274A1')
ax.bar(x - 0.2, base_knn['recall'], width, label='Base KNN - Recall', color='#E1812C')
ax.bar(x - 0.1, base_knn['f1-score'], width, label='Base KNN - F1-Score', color='#3A923A')
ax.bar(x + 0.0, optimized_knn['precision'], width, label='Optimized KNN - Precision', color='#3274A1', alpha=0.5)
ax.bar(x + 0.1, optimized_knn['recall'], width, label='Optimized KNN - Recall', color='#E1812C', alpha=0.5)
ax.bar(x + 0.2, optimized_knn['f1-score'], width, label='Optimized KNN - F1-Score', color='#3A923A', alpha=0.5)
ax.set_ylabel('Score', fontsize=12)
ax.set_xlabel('Classes', fontsize=12)
ax.set_title('Comparison of Base KNN vs KNN with Image Processing and Augmentation', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels([classes[i] for i in range(11)], rotation=45, ha='right')
ax.set_ylim(0, 1)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)
ax.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(base_knn['precision']):
    ax.text(i - 0.3, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8, rotation=90)
for i, v in enumerate(base_knn['recall']):
    ax.text(i - 0.2, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8, rotation=90)
for i, v in enumerate(base_knn['f1-score']):
    ax.text(i - 0.1, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8, rotation=90)
    
for i, v in enumerate(optimized_knn['precision']):
    ax.text(i + 0.0, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8, rotation=90)
for i, v in enumerate(optimized_knn['recall']):
    ax.text(i + 0.1, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8, rotation=90)
for i, v in enumerate(optimized_knn['f1-score']):
    ax.text(i + 0.2, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.savefig('knn_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# CNN plot
# get data
epochs = list(range(10))  # 0-9 epochs
cnn_sgd = [59.67, 74.99, 35.00, 79.71, 79.31, 82.32, 83.15, 83.56, 83.98, 82.61]
cnn_adam = [78.71, 82.08, 80.36, 84.30, 84.81, 85.81, 86.01, 85.44, 85.76, 86.50]
cnn_sgd_aug = [59.63, 72.57, 76.90, 62.15, 79.22, 80.58, 83.22, 82.64, 82.89, 83.52]
cnn_adam_aug = [81.98, 85.37, 84.51, 76.87, 80.59, 87.34, 86.70, 83.69, 86.44, 87.36]

# plot figure
plt.figure(figsize=(12, 8))
plt.plot(epochs, cnn_sgd, marker='o', linewidth=2, label='CNN with SGD')
plt.plot(epochs, cnn_adam, marker='s', linewidth=2, label='CNN with Adam')
plt.plot(epochs, cnn_sgd_aug, marker='^', linewidth=2, label='CNN with SGD + Augmentation')
plt.plot(epochs, cnn_adam_aug, marker='D', linewidth=2, label='CNN with Adam + Augmentation')
plt.axhline(y=cnn_sgd[-1], color='C0', linestyle='--', alpha=0.5)
plt.axhline(y=cnn_adam[-1], color='C1', linestyle='--', alpha=0.5)
plt.axhline(y=cnn_sgd_aug[-1], color='C2', linestyle='--', alpha=0.5)
plt.axhline(y=cnn_adam_aug[-1], color='C3', linestyle='--', alpha=0.5)
plt.text(9.1, cnn_sgd[-1], f'{cnn_sgd[-1]}%', verticalalignment='center')
plt.text(9.1, cnn_adam[-1], f'{cnn_adam[-1]}%', verticalalignment='center')
plt.text(9.1, cnn_sgd_aug[-1], f'{cnn_sgd_aug[-1]}%', verticalalignment='center')
plt.text(9.1, cnn_adam_aug[-1], f'{cnn_adam_aug[-1]}%', verticalalignment='center')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Test Accuracy (%)', fontsize=12)
plt.title('CNN Models Comparison - Test Accuracy by Epoch', fontsize=14)
plt.xlim(-0.5, 10)
plt.ylim(30, 95)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='lower right', fontsize=10)
plt.annotate('SGD drop in epoch 2', 
             xy=(2, 35), 
             xytext=(2.5, 45),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
             fontsize=9)

plt.annotate('Adam + Aug. best overall\nperformance (87.36%)', 
             xy=(9, 87.36), 
             xytext=(6, 92),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
             fontsize=9)

plt.annotate('Adam consistently\noutperforms SGD', 
             xy=(5, 85), 
             xytext=(2, 90),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
             fontsize=9)

plt.xticks(epochs)

# table with final metrics
model_names = ['CNN with SGD', 'CNN with Adam', 'CNN with SGD + Aug.', 'CNN with Adam + Aug.']
final_accuracies = [cnn_sgd[-1], cnn_adam[-1], cnn_sgd_aug[-1], cnn_adam_aug[-1]]
training_times = [210.326, 244.666, 272.922, 270.369]

table_data = []
for i in range(len(model_names)):
    table_data.append([model_names[i], f"{final_accuracies[i]}%", f"{training_times[i]:.1f}s"])

plt.table(cellText=table_data,
          colLabels=['Model', 'Final Accuracy', 'Training Time'],
          loc='bottom',
          bbox=[0, -0.35, 1, 0.2])

plt.subplots_adjust(bottom=0.25)
plt.savefig('cnn_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
classes = {
    0: "Bladder", 
    1: "Femur-left", 
    2: "Femur-right", 
    3: "Heart", 
    4: "Kidney-left", 
    5: "Kidney-right", 
    6: "Liver", 
    7: "Lung-left", 
    8: "Lung-right", 
    9: "Pancreas", 
    10: "Spleen"
}

# best KNN model data (KNN with image processing and augmentation)
best_knn = {
    'class': list(range(11)),
    'precision': [0.47, 0.65, 0.78, 0.90, 0.63, 0.76, 0.88, 0.93, 0.98, 0.56, 0.45],
    'recall': [0.67, 0.58, 0.57, 0.80, 0.61, 0.59, 0.97, 0.97, 0.87, 0.46, 0.64],
    'f1-score': [0.55, 0.61, 0.66, 0.85, 0.62, 0.66, 0.92, 0.95, 0.92, 0.50, 0.53],
    'accuracy': 0.77,
    'training_time': 272.922  # This is actually CNN training time, but using for comparison
}

# best CNN model data (CNN with Adam optimizer and augmentation)
best_cnn = {
    'accuracy': 87.36,  # Final test accuracy
    'training_time': 270.369,  # Training time in seconds
    # Epoch-wise test accuracy (for learning curve)
    'test_accuracy': [81.98, 85.37, 84.51, 76.87, 80.59, 87.34, 86.70, 83.69, 86.44, 87.36]
}

# create figure
fig = plt.figure(figsize=(18, 10))
gs = GridSpec(2, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
model_names = ['Best KNN\n(with Image Processing)', 'Best CNN\n(with Adam + Augmentation)']
accuracies = [best_knn['accuracy'] * 100, best_cnn['accuracy']]  # Convert KNN accuracy to percentage
colors = ['#3274A1', '#E1812C']
bars = ax1.bar(model_names, accuracies, color=colors, alpha=0.7)
ax1.set_ylabel('Test Accuracy (%)', fontsize=12)
ax1.set_title('Overall Accuracy Comparison', fontsize=14)
ax1.set_ylim(0, 100)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 1,
            f'{height:.2f}%', ha='center', va='bottom', fontsize=12)

# add KNN and CNN learning curve 
ax2 = fig.add_subplot(gs[0, 1:])
epochs = range(10)
ax2.plot(epochs, best_cnn['test_accuracy'], marker='o', linewidth=2, color='#E1812C', 
         label='CNN with Adam + Augmentation')
ax2.axhline(y=best_knn['accuracy'] * 100, color='#3274A1', linestyle='--', 
            label=f'KNN Best Accuracy ({best_knn["accuracy"] * 100:.2f}%)')

ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Test Accuracy (%)', fontsize=12)
ax2.set_title('CNN Learning Curve vs KNN Best Performance', fontsize=14)
ax2.set_xticks(epochs)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(loc='lower right')

plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.subplots_adjust(hspace=0.3)

fig.suptitle('Comparison of Best KNN vs Best CNN Models', fontsize=16, y=0.98)
plt.savefig('best_models_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
